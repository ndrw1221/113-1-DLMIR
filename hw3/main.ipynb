# Plot Losses

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
losses = np.load("checkpoints/training_loss.npy")

# Plot the losses
plt.figure(figsize=(10, 6))
plt.plot(losses, marker=".")
plt.grid()
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss")

# Annotate each point with its corresponding loss value
for epoch, loss in enumerate(losses):
    if epoch % 10 == 0:
        plt.text(epoch, loss, f"{loss:.2f}", ha="center", va="bottom", fontsize=8)

plt.show()
```
