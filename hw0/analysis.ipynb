```python
# Load JSON data from your metadata file
with open(TRAIN_PATH / "examples.json", "r") as f:
    data = json.load(f)

# Extract instrument families from the data
instrument_families = [item["instrument_family_str"] for item in data.values()]

# Count occurrences of each instrument family
counts = Counter(instrument_families)

# Define all instrument families (ensure all 11 are included)
instrument_families_list = [
    "bass",
    "brass",
    "flute",
    "guitar",
    "keyboard",
    "mallet",
    "organ",
    "reed",
    "string",
    "synth_lead",
    "vocal",
]

# Ensure counts for all instrument families (fill missing with zero)
counts_complete = {family: counts.get(family, 0) for family in instrument_families_list}

# Plot the distribution using matplotlib
plt.figure(figsize=(10, 6))
plt.bar(counts_complete.keys(), counts_complete.values())
plt.xticks(rotation=45, ha="right")
plt.xlabel("Instrument Family")
plt.ylabel("Count")
plt.title("NSynth subtrain")
plt.tight_layout()
plt.show()
```

# Bass, Guitar, and Keyboard

```python
file_paths = [
    "task1/selected_wavs/bass_electronic_018-040-050.wav",
    "task1/selected_wavs/bass_electronic_018-051-100.wav",
    "task1/selected_wavs/bass_electronic_018-060-075.wav",
    "task1/selected_wavs/guitar_acoustic_030-040-127.wav",
    "task1/selected_wavs/guitar_acoustic_030-051-127.wav",
    "task1/selected_wavs/guitar_acoustic_030-060-075.wav",
    # "task1/selected_wavs/organ_electronic_001-040-050.wav",
    # "task1/selected_wavs/organ_electronic_001-051-075.wav",
    # "task1/selected_wavs/organ_electronic_001-060-127.wav",
    "/home/ndrw1221/nas/nsynth-dataset/nsynth-test/audio/keyboard_electronic_001-033-025.wav",
    "/home/ndrw1221/nas/nsynth-dataset/nsynth-test/audio/keyboard_electronic_001-040-075.wav",
    "/home/ndrw1221/nas/nsynth-dataset/nsynth-test/audio/keyboard_electronic_001-061-075.wav",
]

file_names = [
    "bass electronic 040",
    "bass electronic 051",
    "bass electronic 060",
    "guitar acoustic 040",
    "guitar acoustic 051",
    "guitar acoustic 060",
    # "organ electronic 040",
    # "organ electronic 051",
    # "organ electronic 060",
    "keyboard electronic 033",
    "keyboard electronic 040",
    "keyboard electronic 061",
]
```

```python
# Assuming file_paths and file_names are defined as before
fig, axes = plt.subplots(3, 3, figsize=(15, 10))

# Loop through each file and its corresponding subplot
for i, wav_file in enumerate(file_paths):
    # Load the wav file using librosa
    waveform, sample_rate = librosa.load(wav_file, sr=None)

    # Create the Mel Spectrogram using librosa
    mel_spectrogram = librosa.feature.melspectrogram(
        y=waveform, sr=sample_rate, n_fft=2048, hop_length=512, power=2.0
    )

    # Convert the Mel Spectrogram to decibels
    mel_spec_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

    # Select the corresponding axis (subplot)
    ax = axes[i // 3, i % 3]

    # Plot the Mel spectrogram in the subplot
    im = ax.imshow(mel_spec_db, origin="lower", aspect="auto")

    ax.set_title(f"{file_names[i]}")
    ax.set_xlabel("Time")
    ax.set_ylabel("Mel Frequency")

# Create a colorbar on the right side of the figure that applies to all subplots
fig.subplots_adjust(right=0.85)  # Adjust space on the right for the colorbar
cbar_ax = fig.add_axes(
    [0.87, 0.15, 0.02, 0.7]
)  # Adjust the values to correctly place the colorbar
fig.colorbar(im, cax=cbar_ax).set_label("Amplitude (dB)")  # Add the colorbar and label

# Adjust layout to make space between subplots
plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust the rect to avoid overlap
plt.show()
```

# Log or No Log

```python
file_paths = [
    "task1/selected_wavs/bass_electronic_018-040-050.wav",
    "task1/selected_wavs/guitar_acoustic_030-060-075.wav",
    "/home/ndrw1221/nas/nsynth-dataset/nsynth-test/audio/keyboard_electronic_001-040-075.wav",
]

file_names = [
    "bass electronic 040",
    "guitar acoustic 060",
    "keyboard electronic 040",
]
```

```python
# Assuming file_paths and file_names are already defined

fig, axes = plt.subplots(
    2, 3, figsize=(18, 8)
)  # 2 rows (for scales), 3 columns (for instruments)

# Loop through each instrument (3 instruments)
for i in range(3):
    # Load the wav file for each instrument
    wav_file = file_paths[i]
    waveform, sample_rate = librosa.load(wav_file, sr=None)

    # Create the Mel Spectrogram using librosa
    mel_spectrogram = librosa.feature.melspectrogram(
        y=waveform, sr=sample_rate, n_fft=2048, hop_length=512, power=2.0
    )

    # Plot the Mel spectrogram (without log scale) in the first row
    ax = axes[0, i]  # First row, ith column

    # Plot the Mel spectrogram (original scale)
    im = ax.imshow(mel_spectrogram, origin="lower", aspect="auto")
    ax.set_title(f"{file_names[i]} (Mel Spectrogram)")
    ax.set_xlabel("Time")
    ax.set_ylabel("Mel Frequency")

    # Convert the Mel Spectrogram to decibels (log scale)
    mel_spec_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

    # Plot the log-scaled Mel spectrogram in the second row
    ax_log = axes[1, i]  # Second row, ith column

    # Plot the log-scaled Mel spectrogram
    im_log = ax_log.imshow(mel_spec_db, origin="lower", aspect="auto")
    ax_log.set_title(f"{file_names[i]} (Log Mel Spectrogram)")
    ax_log.set_xlabel("Time")
    ax_log.set_ylabel("Mel Frequency")

# Create a colorbar on the right side of the figure that applies to all subplots
fig.subplots_adjust(right=0.85)  # Adjust space on the right for the colorbar
cbar_ax = fig.add_axes(
    [0.87, 0.15, 0.02, 0.7]
)  # Adjust the values to correctly place the colorbar
fig.colorbar(im_log, cax=cbar_ax).set_label(
    "Amplitude (dB)"
)  # Add the colorbar and label

# Adjust layout to make space between subplots
plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust the rect to avoid overlap
plt.show()
```
