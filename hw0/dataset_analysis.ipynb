```python
from pathlib import Path
import json
import matplotlib.pyplot as plt
from collections import Counter
```

```python
BASE_PATH = Path("/home/ndrw1221/nas/nsynth-dataset")
TRAIN_PATH = BASE_PATH / "nsynth-subtrain"
VALID_PATH = BASE_PATH / "nsynth-valid"
TEST_PATH = BASE_PATH / "nsynth-test"
```

## Class Distribution

```python
# Load JSON data from your metadata file
with open(TRAIN_PATH / "examples.json", "r") as f:
    data = json.load(f)

# Extract instrument families from the data
instrument_families = [item["instrument_family_str"] for item in data.values()]

# Count occurrences of each instrument family
counts = Counter(instrument_families)

# Define all instrument families (ensure all 11 are included)
instrument_families_list = [
    "bass",
    "brass",
    "flute",
    "guitar",
    "keyboard",
    "mallet",
    "organ",
    "reed",
    "string",
    "synth_lead",
    "vocal",
]

# Ensure counts for all instrument families (fill missing with zero)
counts_complete = {family: counts.get(family, 0) for family in instrument_families_list}

# Plot the distribution using matplotlib
plt.figure(figsize=(10, 6))
plt.bar(counts_complete.keys(), counts_complete.values())
plt.xticks(rotation=45, ha="right")
plt.xlabel("Instrument Family")
plt.ylabel("Count")
plt.title("NSynth subtrain")
plt.tight_layout()
plt.show()
```
